## Yield Curve Steepening

A "steepening" occurs when long-term interest rates move more
than short-term rates.

Economically, this can reflect:
- Higher long-run inflation expectations
- Stronger growth outlook
- Rising term premia

We simulate this by increasing long-term rates
and slightly decreasing short-term rates.


```python
def apply_steepener(tenors, rates, size_bp=25):
    """
    Apply a steepening shock to a yield curve.

    Parameters:
    tenors  : array of maturities (years)
    rates   : array of interest rates
    size_bp : size of steepening in basis points

    Returns:
    shocked_rates : new array of rates after steepening
    """
    size = size_bp / 10000  # convert bp to decimal

    # Normalise tenors so short = -1, long = +1
    weights = (tenors - tenors.mean()) / (tenors.max() - tenors.min())

    shocked_rates = rates + size * weights
    return shocked_rates
```

```python
steepened_rates = apply_steepener(tenors, rates, size_bp=25)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(tenors, rates, marker="o", label="Original Curve")
plt.plot(tenors, steepened_rates, marker="o", linestyle="--", label="Steepened Curve")
plt.xlabel("Maturity (years)")
plt.ylabel("Interest Rate")
plt.title("Yield Curve Steepening Scenario")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def apply_curve_shock(tenors, rates, shock_type, size_bp=25, pivot=5):
    """
    Apply a structured yield curve deformation.

    shock_type:
        'steepen'   -> long end up, short end down
        'flatten'   -> short end up, long end down
        'twist'     -> rotate around a pivot maturity
        'butterfly' -> middle moves relative to ends
    """
    size = size_bp / 10000  # bp to decimal

    tenors = np.array(tenors)
    rates = np.array(rates)

    # Normalised position along the curve
    pos = (tenors - tenors.mean()) / (tenors.max() - tenors.min())

    if shock_type == "steepen":
        shock = size * pos

    elif shock_type == "flatten":
        shock = -size * pos

    elif shock_type == "twist":
        shock = size * np.sign(tenors - pivot) * np.abs(pos)

    elif shock_type == "butterfly":
        shock = size * (-(pos ** 2) + np.mean(pos ** 2))

    else:
        raise ValueError("Unknown shock type")

    return rates + shock
```

```python
shock_types = ["steepen", "flatten", "twist", "butterfly"]

plt.figure(figsize=(10, 6))

for shock in shock_types:
    shocked_rates = apply_curve_shock(tenors, rates, shock, size_bp=25)
    plt.plot(tenors, shocked_rates, marker="o", label=shock.capitalize())

plt.plot(tenors, rates, color="black", linewidth=2, label="Original")
plt.xlabel("Maturity (years)")
plt.ylabel("Interest Rate")
plt.title("Yield Curve Deformations")
plt.legend()
plt.grid(True)
plt.show()
```

# What is a bond?
A bond is just a sequence of fixed cash payments in the future.
Its price today depends on interest rates — higher rates mean lower prices.

## Bond Pricing from the Yield Curve

A bond pays fixed cash flows in the future.
To find its price today, we:

1. Take each future cash payment
2. Discount it using the yield curve
3. Add them up

Higher interest rates → lower present value
Lower interest rates → higher present value


```python
# Simple fixed-rate bond
maturity = 5        # years
coupon_rate = 0.04  # 4%
face_value = 100
payments_per_year = 1
```

```python
def bond_cashflows(maturity, coupon_rate, face, freq=1):
    times = np.arange(1, maturity * freq + 1) / freq
    cashflows = np.full_like(times, face * coupon_rate / freq, dtype=float)
    cashflows[-1] += face
    return times, cashflows

times, cashflows = bond_cashflows(maturity, coupon_rate, face_value)
times, cashflows
```

```python
def discount_factor(t, tenors, rates):
    r = np.interp(t, tenors, rates)
    return np.exp(-r * t)

def bond_price(times, cashflows, tenors, rates):
    dfs = np.array([discount_factor(t, tenors, rates) for t in times])
    return np.sum(cashflows * dfs)

price_base = bond_price(times, cashflows, tenors, rates)
price_base
```

```python
price_steep = bond_price(times, cashflows, tenors, steepened_rates)
price_steep
```

```python
def bump_curve(rates, bump_bp):
    """
    Parallel shift to the whole curve by bump_bp (basis points).
    +1 bp = +0.0001 in decimal.
    """
    return rates + (bump_bp / 10000)
```

```python
bump_bp = 1  # 1 basis point

rates_up = bump_curve(rates, bump_bp)
rates_down = bump_curve(rates, -bump_bp)

price_up = bond_price(times, cashflows, tenors, rates_up)
price_down = bond_price(times, cashflows, tenors, rates_down)

price_base, price_up, price_down
```

```python
dy = bump_bp / 10000  # 1bp in decimal

# Effective (modified) duration via central difference
eff_duration = (price_down - price_up) / (2 * price_base * dy)

# DV01: price change for +1bp move (positive number usually reported)
# If rates go UP, price usually goes DOWN, so DV01 is often shown as a positive risk number.
dv01 = price_base - price_up

# Convexity via central difference
convexity = (price_up + price_down - 2 * price_base) / (price_base * dy**2)

eff_duration, dv01, convexity
```

## Risk metrics: Duration, DV01, Convexity

We compute these using a small parallel bump (+/- 1bp) to the entire curve.

- Duration: sensitivity of price to yield changes
- DV01: currency impact for a 1bp move
- Convexity: how duration itself changes for bigger moves


```python
def bump_curve(rates, bump_bp):
    """Parallel bump to all curve points."""
    return rates + bump_bp / 10000

def risk_metrics_parallel(times, cashflows, tenors, rates, bump_bp=1.0):
    """
    Compute Effective Duration, DV01, Convexity using +/- bump.
    """
    p0 = bond_price(times, cashflows, tenors, rates)
    p_up = bond_price(times, cashflows, tenors, bump_curve(rates, +bump_bp))
    p_dn = bond_price(times, cashflows, tenors, bump_curve(rates, -bump_bp))

    dy = bump_bp / 10000

    eff_duration = (p_dn - p_up) / (2 * p0 * dy)
    dv01 = (p_dn - p_up) / 2  # per 1bp (because bump_bp is 1bp by default)
    convexity = (p_up + p_dn - 2 * p0) / (p0 * dy**2)

    return {
        "price": float(p0),
        "price_up": float(p_up),
        "price_down": float(p_dn),
        "duration": float(eff_duration),
        "dv01": float(dv01),
        "convexity": float(convexity),
    }
```

```python
base_risk = risk_metrics_parallel(times, cashflows, tenors, rates, bump_bp=1.0)
base_risk
```

```python
import numpy as np

def normalize_to_minus1_plus1(x):
    """Scale x so min -> -1 and max -> +1."""
    x = np.asarray(x, dtype=float)
    return 2 * (x - x.min()) / (x.max() - x.min()) - 1

def apply_steepener(tenors, rates, size_bp=25):
    """
    Long end up more than short end.
    """
    w = normalize_to_minus1_plus1(tenors)  # short=-1, long=+1
    return rates + (size_bp/10000) * w

def apply_flattener(tenors, rates, size_bp=25):
    """
    Opposite of steepener.
    """
    w = normalize_to_minus1_plus1(tenors)
    return rates - (size_bp/10000) * w

def apply_twist(tenors, rates, size_bp=25, pivot=5.0):
    """
    Twist around a pivot tenor:
    - below pivot moves one way, above pivot moves the other way.
    """
    t = np.asarray(tenors, dtype=float)
    # sign changes around pivot; magnitude increases with distance from pivot
    w = (t - pivot) / (t.max() - t.min())
    return rates + (size_bp/10000) * w

def apply_butterfly(tenors, rates, size_bp=25, belly=5.0, width=3.0):
    """
    Butterfly: belly (middle) moves most.
    We'll use a smooth hump around 'belly'.
    """
    t = np.asarray(tenors, dtype=float)
    # Gaussian hump centered at belly
    w = np.exp(-0.5 * ((t - belly) / width)**2)
    # normalize so max = 1
    w = w / w.max()
    return rates + (size_bp/10000) * w
```

```python
import pandas as pd

def run_scenario(name, shocked_rates):
    r = risk_metrics_parallel(times, cashflows, tenors, shocked_rates, bump_bp=1.0)
    r["scenario"] = name
    r["pnl_vs_base"] = r["price"] - base_risk["price"]
    return r

scenarios = []

# Base
scenarios.append(run_scenario("Base", rates))

# Shape moves
scenarios.append(run_scenario("Steepener +25bp", apply_steepener(tenors, rates, 25)))
scenarios.append(run_scenario("Flattener +25bp", apply_flattener(tenors, rates, 25)))
scenarios.append(run_scenario("Twist +25bp (pivot=5y)", apply_twist(tenors, rates, 25, pivot=5.0)))
scenarios.append(run_scenario("Butterfly +25bp (belly=5y)", apply_butterfly(tenors, rates, 25, belly=5.0, width=3.0)))

df = pd.DataFrame(scenarios)[
    ["scenario", "price", "pnl_vs_base", "duration", "dv01", "convexity"]
].round(6)

df
```

## Curve moves we simulate

- **Steepening**: long-term rates move up more than short-term rates (curve tilts up).
- **Flattening**: long-term rates move up less (or down more) than short-term rates (curve becomes less tilted).
- **Twist**: one end rises while the other falls (like rotating around a middle point).
- **Butterfly**: the middle moves differently than both ends (creates a "hump" or "dip").

These are common macro/rates scenarios used for stress testing and intuition-building.


```python
import numpy as np

def _normalised_slope_weights(tenors):
    """Short end ~ -0.5, long end ~ +0.5 (smooth tilt)."""
    return (tenors - tenors.mean()) / (tenors.max() - tenors.min())

def apply_steepener(tenors, rates, size_bp=25):
    size = size_bp / 10000
    w = _normalised_slope_weights(tenors)
    return rates + size * w

def apply_flattener(tenors, rates, size_bp=25):
    # Exact opposite of steepener
    size = size_bp / 10000
    w = _normalised_slope_weights(tenors)
    return rates - size * w

def apply_twist(tenors, rates, size_bp=25, pivot_year=5):
    """
    Twist around a pivot maturity.
    Rates below pivot move one way; above pivot move the other way.
    """
    size = size_bp / 10000

    # Create weights that are negative left of pivot, positive right of pivot
    w = (tenors - pivot_year) / (tenors.max() - tenors.min())
    return rates + size * w

def apply_butterfly(tenors, rates, size_bp=25, center_year=5):
    """
    Butterfly: the middle moves opposite to the ends.
    This creates a hump/dip shape.
    """
    size = size_bp / 10000

    # Distance from center: small in the middle, large at ends
    d = np.abs(tenors - center_year)
    d = d / d.max()  # 0 at center, 1 at far ends

    # We want middle weight positive, ends negative (or vice versa)
    w = (1 - d) - (1 - d).mean()
    return rates + size * w
```

```python
def run_scenarios(tenors, base_rates, times, cashflows, scenario_specs):
    """
    scenario_specs: dict of {scenario_name: shocked_rates_array}
    returns: list of dict rows (scenario, price, pnl_vs_base)
    """
    base_price = bond_price(times, cashflows, tenors, base_rates)
    rows = []

    for name, shocked_rates in scenario_specs.items():
        p = bond_price(times, cashflows, tenors, shocked_rates)
        rows.append({
            "scenario": name,
            "price": float(p),
            "pnl_vs_base": float(p - base_price)
        })

    return base_price, rows

# Build scenario curves
scenario_curves = {
    "Base": rates,
    "Steepener (+25bp tilt)": apply_steepener(tenors, rates, size_bp=25),
    "Flattener (-25bp tilt)": apply_flattener(tenors, rates, size_bp=25),
    "Twist (+25bp around 5Y)": apply_twist(tenors, rates, size_bp=25, pivot_year=5),
    "Butterfly (+25bp mid vs ends)": apply_butterfly(tenors, rates, size_bp=25, center_year=5),
}

price_base, scenario_rows = run_scenarios(tenors, rates, times, cashflows, scenario_curves)

price_base, scenario_rows[:2]
```

```python
def run_scenarios_with_risk(
    tenors,
    base_rates,
    times,
    cashflows,
    scenario_specs,
    bump_bp=1.0
):
    """
    For each scenario:
    - price the bond
    - compute PnL vs base
    - compute duration, DV01, convexity (parallel bump)
    """
    base_price = bond_price(times, cashflows, tenors, base_rates)
    base_risk = risk_metrics_parallel(times, cashflows, tenors, base_rates, bump_bp)

    rows = []

    for name, shocked_rates in scenario_specs.items():
        risk = risk_metrics_parallel(times, cashflows, tenors, shocked_rates, bump_bp)
        rows.append({
            "scenario": name,
            "price": risk["price"],
            "pnl_vs_base": risk["price"] - base_price,
            "duration": risk["duration"],
            "dv01": risk["dv01"],
            "convexity": risk["convexity"],
        })

    return pd.DataFrame(rows)
```

## Curve Deformation Scenarios (Rates Desk Moves)

We will simulate four standard yield-curve moves:

1) Steepener: long-term rates rise more (or short-term fall)
2) Flattener: opposite of steepener
3) Twist: pivot around a chosen maturity (front end vs back end move opposite)
4) Butterfly: the "belly" (mid maturities) moves differently than wings

These are not random shocks — each has an economic interpretation.


```python
import numpy as np

def _normalised_linear_weights(tenors):
    """
    Map tenors to weights in [-1, +1] increasing with maturity.
    """
    tenors = np.asarray(tenors, dtype=float)
    return (tenors - tenors.mean()) / (tenors.max() - tenors.min())

def apply_steepener(tenors, rates, size_bp=25):
    size = size_bp / 10000
    w = _normalised_linear_weights(tenors)
    return np.asarray(rates, dtype=float) + size * w

def apply_flattener(tenors, rates, size_bp=25):
    # exact opposite sign of steepener
    return apply_steepener(tenors, rates, size_bp=-size_bp)

def apply_twist(tenors, rates, size_bp=25, pivot=5):
    """
    Twist around a pivot maturity:
    - below pivot moves one way
    - above pivot moves the opposite way
    """
    tenors = np.asarray(tenors, dtype=float)
    r = np.asarray(rates, dtype=float)
    size = size_bp / 10000
    
    # weight is negative when tenor < pivot, positive when > pivot
    w = tenors - pivot
    w = w / np.max(np.abs(w))  # scale to [-1, +1]
    
    return r + size * w

def apply_butterfly(tenors, rates, size_bp=25, belly=5):
    """
    Butterfly:
    - belly (mid maturities around `belly`) moves one way
    - wings (short + long) move the opposite way
    """
    tenors = np.asarray(tenors, dtype=float)
    r = np.asarray(rates, dtype=float)
    size = size_bp / 10000
    
    # A smooth "hill" around belly using a Gaussian bump
    # bump ~ 1 at belly, ~0 at far wings
    sigma = (tenors.max() - tenors.min()) / 4
    bump = np.exp(-0.5 * ((tenors - belly) / sigma) ** 2)
    
    # Convert bump into butterfly weights:
    # belly positive, wings negative, mean ~ 0
    w = bump - bump.mean()
    w = w / np.max(np.abs(w))  # scale to [-1, +1]
    
    return r + size * w
```

```python
import matplotlib.pyplot as plt

scenarios = {
    "Steepener (+25bp shape)": apply_steepener(tenors, rates, 25),
    "Flattener (+25bp shape)": apply_flattener(tenors, rates, 25),
    "Twist (+25bp, pivot=5y)": apply_twist(tenors, rates, 25, pivot=5),
    "Butterfly (+25bp, belly=5y)": apply_butterfly(tenors, rates, 25, belly=5),
}

plt.figure(figsize=(9, 6))
plt.plot(tenors, rates, marker="o", linewidth=2, label="Base Curve")

for name, shocked in scenarios.items():
    plt.plot(tenors, shocked, marker="o", linestyle="--", label=name)

plt.xlabel("Maturity (years)")
plt.ylabel("Rate")
plt.title("Yield Curve Deformations")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def risk_metrics(times, cashflows, tenors, curve_rates, bump_bp=1):
    """
    Returns:
    price, effective_duration, dv01, convexity
    """
    bump = bump_bp / 10000
    
    p0 = bond_price(times, cashflows, tenors, curve_rates)
    p_up = bond_price(times, cashflows, tenors, np.asarray(curve_rates) + bump)
    p_dn = bond_price(times, cashflows, tenors, np.asarray(curve_rates) - bump)

    # Effective duration (finite difference)
    eff_dur = (p_dn - p_up) / (2 * p0 * bump)

    # DV01 per 1bp
    dv01 = (p_dn - p_up) / 2  # because bump is 1bp already in price units

    # Convexity (finite difference)
    conv = (p_up + p_dn - 2 * p0) / (p0 * (bump ** 2))

    return float(p0), float(eff_dur), float(dv01), float(conv)

risk_rows = []

# Base
p0, dur, dv01, conv = risk_metrics(times, cashflows, tenors, rates, bump_bp=1)
risk_rows.append(("Base", p0, dur, dv01, conv))

# Scenarios
for name, shocked in scenarios.items():
    p0, dur, dv01, conv = risk_metrics(times, cashflows, tenors, shocked, bump_bp=1)
    risk_rows.append((name, p0, dur, dv01, conv))

risk_df = pd.DataFrame(risk_rows, columns=["Scenario", "Price", "EffDuration", "DV01_per1bp", "Convexity"])
risk_df
```

## Key Rate DV01

Parallel DV01 assumes all interest rates move together.
In reality, different parts of the yield curve move independently.

Key Rate DV01 measures:
- how sensitive the bond price is
- to a 1bp move at a specific maturity (e.g. 2Y, 5Y, 10Y),
- while the rest of the curve stays fixed.


```python
def bump_key_rate(tenors, rates, key_tenor, bump_bp=1):
    """
    Bump only one maturity point on the curve by bump_bp.
    """
    rates = np.asarray(rates, dtype=float).copy()
    bump = bump_bp / 10000

    idx = np.argmin(np.abs(tenors - key_tenor))
    rates[idx] += bump
    return rates
```

```python
def key_rate_dv01(times, cashflows, tenors, rates, key_tenor, bump_bp=1):
    """
    DV01 for a single key maturity.
    """
    p0 = bond_price(times, cashflows, tenors, rates)
    p_up = bond_price(
        times,
        cashflows,
        tenors,
        bump_key_rate(tenors, rates, key_tenor, bump_bp)
    )
    return p_up - p0
```

```python
key_tenors = [2, 5, 10]

kr_rows = []

for kt in key_tenors:
    krdv01 = key_rate_dv01(times, cashflows, tenors, rates, kt, bump_bp=1)
    kr_rows.append((f"{kt}Y", krdv01))

kr_df = pd.DataFrame(kr_rows, columns=["KeyRate", "DV01"])
kr_df
```

```python
# --- Sanity check: Key Rate DV01 should (approximately) add up to total DV01 ---

kr_sum = kr_df["DV01"].sum()

print(f"Sum of Key Rate DV01s (raw): {kr_sum:.6f}")
print(f"Total DV01 (duration block):  {dv01:.6f}")

print("\nCompare magnitudes (ignore sign convention):")
print(f"|Sum(KR DV01)| = {abs(kr_sum):.6f}")
print(f"|Total DV01|  = {abs(dv01):.6f}")
print(f"Absolute gap  = {abs(abs(kr_sum) - abs(dv01)):.6f}")

print("\nIf signs differ, it’s usually just convention:")
print("Often DV01 is reported as a positive risk number, even if price falls when rates rise.")
```

```python
def run_risk_report(
    tenors, base_rates,
    maturity=5, coupon_rate=0.04, face_value=100, freq=1,
    bump_bp=1,
    scenario_size_bp=25,
    pivot=5, belly=5
):
    """
    One function that:
    - builds cashflows
    - prices base bond
    - computes Duration/DV01/Convexity (parallel bump)
    - runs curve scenarios (steepener/ flattener / twist / butterfly)
    - returns two tables: scenario pnl + scenario risk
    """

    # 1) cashflows
    times, cashflows = bond_cashflows(maturity, coupon_rate, face_value, freq=freq)

    # 2) base price
    base_price = bond_price(times, cashflows, tenors, base_rates)

    # 3) parallel risk (Duration/DV01/Convexity)
    # bump rates up/down in parallel
    up_rates = base_rates + bump_bp/10000
    dn_rates = base_rates - bump_bp/10000

    p_up = bond_price(times, cashflows, tenors, up_rates)
    p_dn = bond_price(times, cashflows, tenors, dn_rates)

    eff_dur = (p_dn - p_up) / (2 * base_price * (bump_bp/10000))
    dv01 = (p_dn - p_up) / 2 / bump_bp  # per 1bp, per 100 face (since price is per 100 face)
    conv = (p_up + p_dn - 2*base_price) / (base_price * (bump_bp/10000)**2)

    # 4) scenarios
    scen = []
    scen.append(("Base", base_rates))

    scen.append((f"Steepener (+{scenario_size_bp}bp shape)", apply_steepener(tenors, base_rates, size_bp=scenario_size_bp)))
    scen.append((f"Flattener (+{scenario_size_bp}bp shape)", apply_flattener(tenors, base_rates, size_bp=scenario_size_bp)))
    scen.append((f"Twist (+{scenario_size_bp}bp, pivot={pivot}y)", apply_twist(tenors, base_rates, size_bp=scenario_size_bp, pivot=pivot)))
    scen.append((f"Butterfly (+{scenario_size_bp}bp, belly={belly}y)", apply_butterfly(tenors, base_rates, size_bp=scenario_size_bp, belly=belly)))

    # 5) build tables
    scen_rows = []
    risk_rows = []

    for name, r in scen:
        p = bond_price(times, cashflows, tenors, r)
        scen_rows.append({
            "Scenario": name,
            "BasePrice": base_price,
            "ScenarioPrice": p,
            "PnL_vs_Base": p - base_price
        })

        # scenario-specific parallel risk (optional but nice)
        p_up_s = bond_price(times, cashflows, tenors, r + bump_bp/10000)
        p_dn_s = bond_price(times, cashflows, tenors, r - bump_bp/10000)

        eff_dur_s = (p_dn_s - p_up_s) / (2 * p * (bump_bp/10000))
        dv01_s = (p_dn_s - p_up_s) / 2 / bump_bp
        conv_s = (p_up_s + p_dn_s - 2*p) / (p * (bump_bp/10000)**2)

        risk_rows.append({
            "Scenario": name,
            "Price": p,
            "EffDuration": eff_dur_s,
            "DV01_per1bp": dv01_s,
            "Convexity": conv_s
        })

    scen_df = pd.DataFrame(scen_rows)
    risk_df = pd.DataFrame(risk_rows)

    # 6) print base risk (so it feels like a desk report)
    print(f"Base Price: {base_price:.6f}")
    print(f"Effective Duration: {eff_dur:.4f} years")
    print(f"DV01 (per 1bp): {dv01:.6f} per 100 face value")
    print(f"Convexity: {conv:.2f}")

    return scen_df, risk_df
```

```python
scen_df, risk_df = run_risk_report(tenors, rates)

scen_df, risk_df
```

```python
import pandas as pd

def run_full_simulation(tenors, rates, maturity=5, coupon_rate=0.04, face_value=100, key_tenors=(2,5,10)):
    from ycdeform.bond import bond_cashflows
    from ycdeform.engine import simulate_scenarios

    times, cashflows = bond_cashflows(maturity, coupon_rate, face=face_value, freq=1)

    scenarios_df, risk_df, keyrate_df = simulate_scenarios(
        tenors=tenors,
        base_rates=rates,
        times=times,
        cashflows=cashflows,
        key_tenors=list(key_tenors)
    )

    return scenarios_df, risk_df, keyrate_df
```
